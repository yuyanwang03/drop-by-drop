```python
# Read data
daily_economic = pd.read_csv("./data/daily_dataset_economic_activity.csv", encoding='latin1') # Latin encoding because accent marks give errors 
daily_economic.info()

print("")
daily = pd.read_csv("./data/daily_dataset.csv", encoding='latin1')
daily.info()

print("")
general = pd.read_csv("./data/dades_datachallenge.csv", encoding='latin1')
general.info()
```

```python
# Create a histogram for the 'Accumulated consumption (L/day)' column
plt.figure(figsize=(10, 6))
plt.hist(daily_economic['Consum acumulat (L/dia)/Consumo acumulado (L/día)/Accumulated consumption (L/day)'], bins=20, color='blue', edgecolor='black')
plt.title('Histogram of Accumulated Consumption (L/day)')
plt.xlabel('Accumulated Consumption (L/day)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```
